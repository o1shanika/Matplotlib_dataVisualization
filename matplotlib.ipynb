```python
import matplotlib.pyplot as plt

# Assuming x and y are defined
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

```

```python
plt.plot(x, y, color='red', linestyle=':', linewidth=2)
plt.title('Example for Line Plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid('True')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]
```

```python
#_________________________________________subline plots________________________________________plt.subplot()
```

```python
#sub line plots
plt.subplot(1, 2, 1)  # Subplot 1
plt.plot(x, y1, color='g', linestyle=':')
plt.title('Subplot 1')
plt.xlabel('x axis')

plt.subplot(1, 2, 2)  # Subplot 2
plt.plot(x, y2, color='r', linestyle=':')
plt.title('Subplot 2')
plt.xlabel('x axis')

#adjust the space between two subplots
plt.subplots_adjust(wspace=0.4)

plt.tight_layout()  # Optional: Adjust layout for better spacing
plt.show()
```

```python
#_________________________________________bar plots________________________________________plt.bar()
```

```python

students={"Bob":57,"Matt":56,"sam":27}
names = list(students.keys())
ages = list(students.values())

# Plotting the bar chart
plt.bar(names, ages, color='skyblue')

# Adding labels and title
plt.xlabel('Students')
plt.ylabel('Age')
plt.title('Ages of Students')

# Display the plot
plt.show()
```

```python

x=[1,2,3,4,5,6]
y=[10,20,30,40,50,60]

plt.scatter(x,y,marker='*',c='r',s=100)
```

```python
#____________________________________________________histogram_______________________________________________plt.hist()
```

```python
# Example data
data = np.random.randn(1000)  # Generate 1000 random numbers from a normal distribution

# Creating a histogram
plt.hist(data, bins=30, color='blue', edgecolor='black')

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

# Display the plot
plt.show()
```

```python
#____________________________________________________boxplots_______________________________________________plt.boxplot()
```

```python
data1 = np.random.randn(100)  # Generate 100 random numbers from a normal distribution
data2 = np.random.rand(100) * 100  # Generate 100 random numbers from a uniform distribution scaled by 100

# Creating a box plot
plt.boxplot([data1, data2], labels=['Normal Distribution', 'Uniform Distribution'])

# Adding labels and title
plt.ylabel('Value')
plt.title('Box Plot of Different Distributions')

# Display the plot
plt.show()
```
